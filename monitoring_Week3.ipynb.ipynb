```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = pd.read_csv("data/etl_log.csv")
print("ETL Log Loaded. Shape:", df.shape)
print("\nLast few records:\n")
print(df.tail())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="status", palette="viridis")
plt.title("ETL Run Status Overview")
plt.xlabel("ETL Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data=df, x="timestamp", y=df.index, marker="o", color="teal")
plt.title("ETL Run Timeline")
plt.xlabel("Run Timestamp")
plt.ylabel("Run Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
